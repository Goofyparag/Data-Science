```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000] # Chai Point
])

print("==== Zomato Sales Analysis ====")
print(f"\n Sales data shape : ", sales_data.shape)
print(f"\n Sammple data for 1st 3 rest :\n", sales_data[0:3])
```

```python
#total  sales per year 
print(np.sum(sales_data, axis=0))
year_total =  np.sum(sales_data[:,1:], axis=0)
print(year_total)
```

```python
#Minimun Slaes per resturant
min_sales =  np.min(sales_data[:,1:], axis=1)
print(min_sales)
```

```python
#MMaximum Slaes per resturant
max_sales =  np.max(sales_data[:,1:], axis=1)
print(max_sales)
```

```python

# Average sales per resturant
avg_sales = np.mean(sales_data[:,1:], axis=1)
avg_sales2 = np.mean(sales_data[:,1:], axis=0)
print(avg_sales)
print(avg_sales2)
```

```python
array = np.array([1, 2, 3, 4, 5])
cumulative_sum = np.cumsum(array)

print("Original array:", array)
print("Cumulative sum:", cumulative_sum)
```

```python
#Cumalative Sum Sales

cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)


#matplotlib

plt.figure(figsize=(4, 3))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumalative sales across all resturant.")
plt.xlabel("Years")
plt.ylabel("Sales")

plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print(f'Vector Addition: {vector1+vector2}')
print(f'\nVector Multiplication: {vector1*vector2}')

print('\nDot Product: ',np.dot(vector1, vector2))

print('\nPercentile: ',np.percentile(vector1, vector2))

# print('\nQuantile: ',np.quantile[0,4])

# angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
# print('\nAngle of Vectors: ',angle)
```

```python
manual_dot_product = 0
for i in range(len(vector1)):
    manual_dot_product += vector1[i] * vector2[i]

print(f"Manual Dot Product: {manual_dot_product}")
```
