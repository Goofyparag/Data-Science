```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#line plot
X=np.array([5,10,8,9])
Y=np.array([12,30,1,8])
plt.plot(X,Y)
plt.plot(X,Y,marker="1")
plt.show()
plt.plot(X,Y,"o")
#marker, line style, color
#for black color use "k"

```

```python
x1=np.array([0,1,2,3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc="right")
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)


plt.show()
```

```python
plt.figure(figsize=(10, 4))
#fu=iguresize (width, height)
# Generate random data
data = np.random.randn(1000)

# Create the histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Add labels and a title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')


# Display the plot
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, autopct='%1.2f%%', shadow=True, startangle=140)
plt.legend(title="fruits")
plt.figure(figsize=(4, 40))
plt.show()
```
