```python
import numpy as np
data = np.random.normal(loc=0, scale=1, size=1000)

#histplot with KDE
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True, bins=30, color='skyblue', stat='density')
plt.title('Distribution of Sample Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
