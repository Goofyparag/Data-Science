#  Adding & Modifying Columns

```python
df["SALES_PER_ITEM"] = df["TOT_SALES"] / df["PROD_QTY"]  # Create new column
df["DISCOUNT"] = 0.1                                     # Add fixed value column
df["PROD_NAME"] = df["PROD_NAME"].str.upper()             # Modify column values
```

# Grouping & Aggregation

```python
df.groupby("STORE_NBR")["TOT_SALES"].sum()                 # Total sales per store
df.groupby("STORE_NBR").agg({"TOT_SALES": "mean", "PROD_QTY": "sum"})
df.groupby(["STORE_NBR", "PROD_NAME"]).size()              # Count per gr
```

```python
df["DATE"] = pd.to_datetime(df["DATE"], origin="1899-12-30", unit="D")  # Excel date conversion
```

# Merging & Joining

```python
# Example merge with product details
prod_info = pd.DataFrame({
    "PROD_NBR": [94],
    "CATEGORY": ["Snacks"]
})
df = df.merge(prod_info, on="PROD_NBR", how="left")
```

# Pivot Tables

```python
pd.pivot_table(df, values="TOT_SALES", index="STORE_NBR", columns="PROD_NAME", aggfunc="sum")
```

# Exporting Data

```python
df.to_csv("sales_data.csv", index=False)
df.to_excel("sales_data.xlsx", index=False)
```
