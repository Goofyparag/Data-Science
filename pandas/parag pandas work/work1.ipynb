```python
#print('Before sort — top 5 rows:')
display(df.head())
# Sort by LIFESTAGE (ascending) then PREMIUM_CUSTOMER (descending) in-place
# - Primary key: LIFESTAGE (alphabetically A → Z)
# - Secondary key: PREMIUM_CUSTOMER (Premium first, then Mainstream, then Budget)
df.sort_values(['LIFESTAGE', 'PREMIUM_CUSTOMER'], ascending=[True, False], inplace=True)

# Show top rows after sort to verify
print('\nAfter sort — top 5 rows:')
display(df.head())

# Notes:
# - Using inplace=True modifies `df` directly and returns None.
```

```python
# Histogram of PREMIUM_CUSTOMER after mapping to numeric categories
mapping = {'Budget': 0, 'Mainstream': 1, 'Premium': 2}
data_cat = df['PREMIUM_CUSTOMER'].map(mapping).dropna()
plt.figure(figsize=(6,3))
plt.hist(data_cat,bins=[-0.5,0.5,1.5,2.5], color='lightgreen', edgecolor='black')
plt.xticks([0,1,2], ['Budget','Mainstream','Premium'])
plt.xlabel('PREMIUM_CUSTOMER')
plt.ylabel('Count')
plt.title('Histogram of PREMIUM_CUSTOMER ')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show() 
```

```python
s=df['LIFESTAGE'].value_counts()
mylabels = s.index.tolist()
plt.pie(s, labels = mylabels, startangle = 90,autopct='%1.2f%%', shadow=True)
plt.show()
```

mylabels = s.index.tolist()

.tolist() converts these index values into a standard Python list, which is stored in the variable mylabels. These labels will be used for the slices of the pie chart
