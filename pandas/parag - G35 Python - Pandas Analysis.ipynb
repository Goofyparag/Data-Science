```python
pb.rename(columns={'Customer_ID': 'LYLTY_CARD_NBR'}, inplace=True)
td.rename(columns={'LYLTY_CARD_NUMBER': 'LYLTY_CARD_NBR'}, inplace=True)
```

```python
pb.columns = pb.columns.str.strip()
td.columns = td.columns.str.strip()
print(pb.columns.tolist())
print(td.columns.tolist())
```

```python

pb.rename(columns={'Loayality_card_id': 'LYLTY_CARD_NBR'}, inplace=True)

common_customers = set(pb['LYLTY_CARD_NBR']).intersection(set(td['LYLTY_CARD_NBR']))
print("Common Customers:", len(common_customers))
```

```python
# Make sure column names are standardized
pb.rename(columns={'Loayality_card_id': 'LYLTY_CARD_NBR'}, inplace=True)

# Merge
merged_df = td.merge(pb, on='LYLTY_CARD_NBR', how='left')

print(merged_df.head())
print(merged_df.info())
```

```python
sales_by_product = merged_df.groupby('PROD_NAME')['TOT_SALES'].sum().sort_values(ascending=False)
print(sales_by_product.head(10))
```

```python
avg_spend_segment = merged_df.groupby('LIFESTAGE')['TOT_SALES'].mean().sort_values(ascending=False)
print(avg_spend_segment)
```

Sales by Lifestage & Premium Status

```python
pivot_sales = pd.pivot_table(
    merged_df,
    values='TOT_SALES',
    index='LIFESTAGE',
    columns='PREMIUM_CUSTOMER',
    aggfunc='sum',
    fill_value=0
)
print(pivot_sales)
```

Average quantity by product and lifestage

```python
pivot_qty = pd.pivot_table(
    merged_df,
    values='PROD_QTY',
    index='PROD_NAME',
    columns='LIFESTAGE',
    aggfunc='mean',
    fill_value=0
)
print(pivot_qty.head())
```
